## importing packages & getting files & initial settings

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'AppleGothic'

filename = "/Users/nyh211/Desktop/CAU/1-1/일반물리실험/일반물리실험 3차시 (글라이더의 가속도 측정)/일반물리실험3차시(글라이더의 가속도 측정 표).xlsx"
```

## setting fonts

```python
#setting fonts

title_font = {
    'size': 17
} #fontdict of title
fontdict = {
    'size': 11
}
#setting titles
```

## 데이터 추츨 & Default settings

```python
data_sheet = pd.read_excel(filename, sheet_name=None, engine='openpyxl')
data_list = []

x_axis = [1, 2, 3, 4, 5] #default
```

## 추의 질량 변화 그래프 코드 작성

```python
#추의 질량 변화 코드 작성란

for i in range(1, 4): # 질량을 변화해가며 실험하는 과정
    data_list = data_sheet['weight'+str(i)].values.tolist()
    
    mass_of_glider = data_list[1][1]
    mass = data_list[2][1]
    mean_value = [data_list[12][1]]*5 #mean_value of measurement values
    
    y_axis = []
    real = [data_list[7][2]]*5 # repeat 5 times to plot graph
    
    for j in range(len(data_list)):
        print(data_list[j])
        
    print('')
    
    for j in range(7, 12):
        y_axis.append(data_list[j][1])
    
    print(y_axis)
    print('')
    
    #making ticks & scope of x, y axis
    plt.xticks(x_axis)
    plt.yticks(y_axis+real+mean_value)

    #plt.ylim(295, 330)
    
    #mesured values
    plt.plot(x_axis, y_axis, 'ro', label = '39cm')
    plt.plot(x_axis[3:5], y_axis[3:5], 'bo', label = '29cm') #다른 범주 덮어씌움
    plt.plot()
    #--- real value ---#
    plt.plot(x_axis, real, 'g--', label = '참값')
    #--- mean value ---#
    plt.plot(x_axis, mean_value, color = 'black', linestyle = '--', label='평균값')
    #styling
    
    plt.rc('ytick', labelsize=8)
    plt.rc('xtick', labelsize=8)
    
    plt.xlabel('회', fontdict)
    plt.ylabel('가속도($cm/s^{2}$)', fontdict)
    
    plt.text(2.5, max(y_axis), "추의 질량(g): "+str(mass))
    plt.title("추의 질량 변화", title_font)
    plt.grid(True, alpha = 0.3, linestyle='--')
    
    plt.legend(loc='best')
    
    plt.show()
```

## 추의 질량 변화 오차율 그래프 코드

```python
#추의 질량 변화 오차율 그래프 코드 작성란

for i in range(1,4):
    data_list = data_sheet['weight'+str(i)].values.tolist()
    
    mass = data_list[2][1]
    y_axis = []
    
    for i in range(7, 12):
        y_axis.append(abs(data_list[i][3]))
        
    mean_value_long = [sum(y_axis[:3])/3]*3
    mean_value_short = [sum(y_axis[3:5])/2]*2
    
    plt.xticks(x_axis)
    plt.yticks(y_axis+real+mean_value)
    
    plt.rc('ytick', labelsize=8)
    plt.rc('xtick', labelsize=8)
    
    plt.plot(x_axis[:3], y_axis[:3], 'ro', label = '39cm')
    plt.plot(x_axis[3:5], y_axis[3:5], 'bo', label = '29cm')
    
    plt.plot(x_axis[:3], mean_value_long, 'r--', label ='39cm 평균')
    plt.plot(x_axis[3:5], mean_value_short, 'b--', label ='29cm 평균')
    
    plt.title("오차율(%)", title_font)
    
    plt.xlabel('회', fontdict)
    plt.ylabel('오차율(%)', fontdict)
    
    plt.text(4, min(y_axis), "추의 질량(g): "+str(mass))
    
    plt.grid(True, alpha = 0.3, linestyle='--')
    plt.legend(loc='best')
    
    plt.show()
```

## 글라이더의 질량 & 추의 질량 변화 그래프 코드

```python
#글라이더의 질량 & 추의 질량 변화 그래프 코드 작성란

for i in range(1, 4): # 질량을 변화해가며 실험하는 과정
    data_list = data_sheet['glider'+str(i)].values.tolist()
    
    mass_of_glider = data_list[1][1]
    mass = data_list[2][1]
    mean_value = [data_list[12][1]]*5 #mean_value of measurement values
    
    y_axis = []
    real = [data_list[7][2]]*5 # repeat 5 times to plot graph
    
    for j in range(len(data_list)):
        print(data_list[j])
        
    print('')
    
    for j in range(7, 12):
        y_axis.append(data_list[j][1])
    
    print(y_axis)
    print('')
    
    #making ticks & scope of x, y axis
    plt.xticks(x_axis)
    plt.yticks(y_axis+real+mean_value)

    #plt.ylim(295, 330)
    
    #mesured values
    plt.plot(x_axis, y_axis, 'ro', label = '39cm')
    plt.plot(x_axis[3:5], y_axis[3:5], 'bo', label = '29cm') #다른 범주 덮어씌움
    plt.plot()
    #--- real value ---#
    plt.plot(x_axis, real, 'g--', label = '참값')
    #--- mean value ---#
    plt.plot(x_axis, mean_value, color = 'black', linestyle = '--', label='평균값')
    #styling
    
    plt.rc('ytick', labelsize=8)
    plt.rc('xtick', labelsize=8)
    
    plt.xlabel('회', fontdict)
    plt.ylabel('가속도($cm/s^{2}$)', fontdict)
    
    plt.text(2.5, max(y_axis), "추의 질량(g): "+str(mass))
    plt.text(2.3, max(y_axis)-1, "글라이더의 질량(g): "+str(mass_of_glider))
    
    plt.title("추&글라이더의 질량 변화", title_font)
    plt.grid(True, alpha = 0.3, linestyle='--')
    
    plt.legend(loc='best')
    
    plt.show()
```

## 글라이더의 질량 & 추의 질량 변화 오차율 그래프 코드

```python
#글라이더의 질량 & 추의 질량 변화 오차율 그래프 코드 작성란

for i in range(1,4):
    data_list = data_sheet['glider'+str(i)].values.tolist()
    
    mass_of_glider = data_list[1][1]
    mass = data_list[2][1]
    
    y_axis = []
    
    for i in range(7, 12):
        y_axis.append(abs(data_list[i][3]))
        
    mean_value_long = [sum(y_axis[:3])/3]*3
    mean_value_short = [sum(y_axis[3:5])/2]*2
    
    plt.xticks(x_axis)
    plt.yticks(y_axis+real+mean_value)
    
    plt.rc('ytick', labelsize=8)
    plt.rc('xtick', labelsize=8)
    
    plt.plot(x_axis[:3], y_axis[:3], 'ro', label = '39cm')
    plt.plot(x_axis[3:5], y_axis[3:5], 'bo', label = '29cm')
    
    plt.plot(x_axis[:3], mean_value_long, 'r--', label ='39cm 평균')
    plt.plot(x_axis[3:5], mean_value_short, 'b--', label ='29cm 평균')
    
    plt.title("오차율(%)", title_font)
    
    plt.xlabel('회', fontdict)
    plt.ylabel('오차율(%)', fontdict)
    
    plt.text(4, min(y_axis)+0.2, "추의 질량(g): "+str(mass))
    plt.text(3.57, min(y_axis), "글라이더의 질량(g):"+str(mass_of_glider))
    
    plt.grid(True, alpha = 0.3, linestyle='--')
    plt.legend(loc='best')
    
    plt.show()
```
